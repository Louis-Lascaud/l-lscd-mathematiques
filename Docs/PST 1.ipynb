Exponentiation rapide (pour le chiffrement et le déchiffrement)

```python
def expo_rapide(x,n):
    '''
    Entrées :
        x : nombre complexe
        n : entier naturel non nul
    Sortie : x^n
    '''
    if n==1:
        return x
    elif n%2==0:
        return expo_rapide(x^2,n/2)
    else:
        return x*expo_rapide(x^2,(n-1)/2)
```

```python
expo_rapide(2,4)
```

Algorithme d'Euclide étendu

```python
def euclide_etendu(a,b):
    '''
    Entrées :
        a, b : entiers naturels avec b différent de 0
    Sortie :
        triplet (d, u, v) : d = pgcd(a, b) et au + bv = d
    '''
    Q = []
    q = 0
    r = 1
    #descente de l'algorithme d'Euclide
    while r!=0:
        q = a//b
        Q+=[q]
        r = a%b
        a = b
        d = b
        b = r
    #remontée (matricielle) de l'algorithme d'Euclide
    A = Matrix([[1,0],[0,1]])
    for q in Q:
        A=Matrix([[0,1],[1,-q]])*A
    u = A[0][0]
    v = A[0][1]
    return (d,u,v)
```

```python
euclide_etendu(7,6)
```

Inversion modulaire

```python
def inverse_mod(x,n):
    '''
    Entrées :
        n : entier naturel non nul
        x : entier naturel premier avec n
    Sortie :
        u : inverse de x modulo n
    '''
    return euclide_etendu(x,n)[1]
```

Isomorphisme réciproque des restes chinois

```python
def inversion_chinoise_naive(C,N):
    '''
    Entrées :
        N : liste de taille r des "modules"
        C : liste de taille r des données
    Sortie :
        x : entier modulo tel que x = C_i dans Z/(N_i)Z pour tout i entre 1 et r
    '''
    r=len(N)
    n=prod(N)
    Q = [n/N[i] for i in range(r)]
    I = [inverse_mod(Q[i],N[i]) for i in range(r)]
    sol = 0
    for i in range(r):
        sol+=C[i]*I[i]*Q[i]
    return sol
```

```python
def inversion_chinoise(C,N):
    '''
    Entrées :
        N : liste de taille r des "modules"
        C : liste de taille r des données
    Sortie :
        x : entier modulo n = prod(N) tel que x = C_i dans Z/(N_i)Z pour tout i entre 1 et r
    '''
    r=len(N)
    n=prod(N)
    Q = [n/N[i] for i in range(r)]
    I = [inverse_mod(Q[i],N[i]) for i in range(r)]
    sol = 0
    for i in range(r):
        sol+=C[i]*I[i]*Q[i]
    A=IntegerModRing(n)
    return A(sol)
```

```python
inversion_chinoise([3,4,5],[17,11,6])
```

Extraction de racines dans N

```python
def racine_nieme(x,e,t):
    '''
    Entrées :
        x : entier naturel
        e : entier naturel
    Sortie :
        y : entier naturel tel que y^e = x
    '''
    N = lambda y : y - (y^e-x)/(e*y^(e-1))
    for i in range(6):
        t = N(t)
    R = RR(t)
    E = floor(R)
    if abs(R-E)>1/2:
        return E+1
    else:
        return E
```

```python
racine_nieme(125,3,42)
```

Notation binaire des entiers

```python
def binaire(n):
    '''
    Entrée :
        n : entier naturel non nul
    Sortie :
        L : liste des coefficients de n écrit en base b, de gauche à droite
    '''
    L=[]
    while n!=0:
        L+=[n%2]
        n=n//2
    return L[::-1]
```
