# Implémentation sous forme polynomiale des motifs et canons

```python
def motif(M):
    '''
    Entrée :
    M : partie finie non vide de N (sous la forme d'un ensemble)
    Sortie :
    Représentation polynomiale du motif M
    '''
    assert M != set()
    R.<x>=ZZ['x']
    P = sum([x^k for k in M])
    return P
```

```python
M = {0, 1, 2}
```

```python
motif(M)
```

```python
def canon(C):
    '''
    Entrée :
    C = (M,D) : canon où :
    M : motif
    D : décalage fini, i. e. partie finie non vide de N (sous la forme d'un ensemble)
    Sortie :
    Représentation polynomiale du canon C = (M, D)
    '''
    M = C[0]
    D = C[1]
    assert M != set()
    assert D != set()
    R.<x>=ZZ['x']
    P = sum([x^(k+d) for k in M for d in D])
    return P
```

```python
M = {0, 2}
D = {0, 1, 4, 5}
C = (M,D)
```

```python
canon(C)
```

```python
def somme(A,B):
    '''
    Entrées :
    A, B : parties finies non vide de N
    Sortie :
    Booléen prenant la valeur [C est un canon direct]
    '''
    assert A != set()
    assert B != set()
    S = {a + b for a in A for b in B}
    return S
```

```python
somme({0,1},{0,1})
```

```python
def is_direct(C):
    '''
    Entrée :
    C : canon fini
    Sortie :
    Booléen prenant la valeur [C est un canon direct]
    '''
    M = C[0]
    D = C[1]
    S = somme(M,D)
    P0 = motif(S)
    P1 = motif(M)
    P2 = decalage(D)
    P3 = P1*P2
    return bool(P0 == P3)
```

```python
def is_direct_rapide(C):
    '''
    Entrée :
    C : canon fini
    Sortie :
    Booléen prenant la valeur [C est un canon direct]
    '''
    M = C[0]
    D = C[1]
    S = somme(M,D)
    P = motif(S)
    n = P.degree()
    booleen = 1
    for i in range(n):
        booleen=booleen*(bool(P[i]==0 or P[i]==1))
    return bool(booleen)
```

```python
is_direct_rapide(CC)
```

# Pavages

```python
S.<z>=ZZ['z']
def quopar(n):
    '''
    Entrée :
    n : entier naturel
    Sortie :
    Anneau des polynômes sur Z en la variable x quotienté par l'idéal (X^n - 1)
    '''
    J = S.ideal(z^n - 1)
    return S.quo(J)
```

```python
quopar(2)(z + 1)
```

```python
quopar(2)(z^2 + 1)
```

```python
def sigma(n):
    return sum([z^i for i in range(n)])
```

```python
sigma(7)
```

```python
def test_pavage(M,D,n):
    '''
    Entrée :
    M : motif
    D : décalage
    n : candidat pour une période
    Sortie :
    Booléen prenant la valeur [(M,D) est un canon périodique de période n]
    '''
    A = quopar(n)
    P = motif(M)
    Q = decalage(D)
    R = P*Q
    S = A(R)
    s = sigma(n)
    sigmapro = A(s)
    return bool(S == sigmapro)
```

```python
test_pavage(M,D,6)
```

```python
test_pavage(M,D,72)
```

# Construction de canons périodiques

```python
def monomes_apparaissant(P):
    '''
    Entrée :
    P : polynôme à coefficients dans Z
    Sortie :
    Ensemble des degrés des monômes non nuls de P
    '''
    n = P.degree()
    L = []
    for i in range(n+1):
        if P[i]!=0:
            L+=[i]
    S = set(L)
    return S
```

```python
monomes_apparaissant(sigma(7))
```

```python
def new_can_per(M,D,n,k):
    '''
    Entrée :
    (M, D) : canon périodique de période n
    k : entier naturel
    Sortie :
    (M', D') : canon périodique de période nk construit selon le Corollaire démontré en classe.
    '''
    PM = motif(M)
    PD = decalage(D)
    Q = sigma(k)*PM(z^k)
    R = PD(z^k)
    Mprime = set(monomes_apparaissant(Q))
    Dprime = set(monomes_apparaissant(R))
    return (Mprime,Dprime)
```
